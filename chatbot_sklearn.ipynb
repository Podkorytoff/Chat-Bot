```python
import pandas as pd
from sklearn import tree
import graphviz
import numpy as np
```

```python
train = pd.read_csv(r"C:\Users\Yan\Desktop\UFABC\Programação estruturada\Projeto Chatbot\dados_final.csv", sep = ",")
y_train = train['CUSTO']
x_train_columns = train.drop(['CUSTO'], axis = 1).columns
x_train = train.drop(['CUSTO'], axis = 1).values
decision_tree = tree.DecisionTreeClassifier(max_depth = 20)
decision_tree.fit(x_train, y_train)
```

```python
dot_data = tree.export_graphviz(decision_tree, out_file=None, feature_names=x_train_columns, filled=True, rounded=True, special_characters=True)
```

```python
graph = graphviz.Source(dot_data)
```
